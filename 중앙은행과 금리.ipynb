# 금리와 은행
* 금리란?<br>
금리는 하나의 가격, 돈이 필요한 사람과 돈을 굴리려는 사람이 만나 결정하는 이자율을 뜻한다<br>
금리 = 현재 경제 상황이 반영된 최소한의 이자율 + 기간을 고려한 신용위험 & 기회비용<br>
<br>
* 실질금리가 중요!<br>
명목금리 = 실질금리 + 물가상승률<br>
하지만 우리는 물가상승률을 정확하게 파악할수 없으므로 실질금리는 추정할 뿐이다<br>
실질금리가 높다는 것은, 나라의 경제활동이 매우 활발하여 금리가 높음에도 불구하고 대출을 통해 어디엔가 투자하고자 하는 사람들이 많다는 것이다<br>
선진국 < 개도국 : 경제가 활발하고 성장률이 높은 나라에서 금리가 높게 형성된다<br>
cf)경제 위기에 실질 금리의 마이너스 상태를 심화시켜, 소비 대출과 투자를 유인하기도 함<br>
<br>
* 금리와 은행(중앙은행과 시중은행)이 중요한 이유<br>
은행은 신중해야 한다. 세상 자원은 한정적이고 수요는 어느순간 몰려들때도 갑자기 사라질수도 있다. 이것은 결국 돈의 총량에 달려있고, 돈은 중앙은해이 발행하고, 시중은해이 확장시킨다. 따라서 은행은 버블을 만들 수 있고, 신용경색도 만들 수 있다. 금리를 조절하여 유동성을 적절히 공급하여야한다 <br>



#1. 한국 중앙은행의 역할과 금리조정의 주요 이슈들

#3. 중국 금리조정의 주요 이슈들
>* 중국은 기준금리보다 지급준비율로 조절 <br> 
>> * 지급준비율로 국가경제를 제어하는 대표적인 나라가 중국으로 중국은 국가가 여러부분을 직접 통제<br>
>> * 금리를 이용하여 투자수요랑 물가 조절을 하는것이 아니다 <br>
>> * 경제 성장률을 위해 국가전략적으로 시중금리는 낮추되, 지준율로 대출총액을 제한하고, 국가가 전략적으로 성장시켜 하는 부분에만 대출<br>
>> * 대출의 기회를 박탈한다는 점에서 불합리하지만 중국이라서 가능하고, 비리가 발생할 가능성이 높음<br> 
> * 장기적인 저금리 정책은 과도한 물가상승 그리고 이에 따른 사회양극화를 일으키지만 중국은 통제때문에 가능함
