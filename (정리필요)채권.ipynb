#2. 한국의 채권시장

>* **채권이란**
>> 채권이란 기업이 자금을 조달하는 두가지 방법의 하나, 기업이 안정되고 성장성이 있다면 주식발행보다 채권발행이 훨씬 유리하다 <br>
그래서 주식으로  자금을 조달하느 유상증자가 안좋은 신호이기도함<br>
>* **채권의종류**
>>* 국가가 발행하는 국채 : 세금보다 쓸 돈이 많을때, 미래 세금을 담보로 발행하는 채권<br>
>>* 중앙은행이 발행하는 통화안정증권(통안채) : 유동성을 조절하는 역할, 정부부채의 일부???<br>
>>* 정부소유 공기업들이 발행한 특수채(공사채) : 한국주택금융공사의 MBS (주택담보대출,보금자리론)<br>
>>* 은행이 발행 은행채 : 대출수요가 더 많을때 채권발행을 통해 조달, 예대율(예금잔액대비대출잔액비율) 규제로 적절히 관리해야한다<br>
>>* 기업이 발행 회사채 <br>
>* **채권가격과 채권금리**<br>
>> 채권가격과 채권금리는 반대로 움직임<br>
> 채권수요 증가 -> 채권가격상승 -> 채권금리하락
>> - 한국의 채권금리 
>>> 국채 30년물 금리가 10년물보다 낮은 유일한 나라이다 : 연기금 때문인데 별로 주 <br>

>* **채권금리**
>> - 위기시 
>>> * 단기채권금리 수요하락 -> 단개채권가결하락 & 단기채권금리상승<br>
>>> * 장기채권금리 수요증가 -> 장기채권가격상승 & 장기채권금리하락<br>
>>> => 스프레드 줄어듬 = 장단기 금리차 축소<br> 
>> - 장기 채권금리는 어느정도가 적정선인가?<br>
>> 장기 채권금리가 하락하면 장단기 금리차 역전걱정, 장기 채권금리가 상승하면 그것대로 불안하다고 함 <br>
>>> 장기 채권금리의 급락은 경기 하락에 대한 우려가 있는것이긴 함  ex)ISM제조업지수가 안좋은것<br>
>>> 하지만 채권금리는 원인이 되는 요소에 대한 시장 참여자들의 해석,투표 정도로 받아 들여야한다<br>
>>> 즉 같은 원인에 대해 금리 상승일 수도, 하락일 수도 있다<br>
>>> 채권 금리로 경기를 판단하기 보단,경기판단 하기 위해서는 특히 환율 및 다른지표들을 보아라<br>
