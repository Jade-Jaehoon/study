# 2008년 양적완화 : 공개시장매입규모를 초단기국채+장기국채까지 확대한 것을 의미
>원래 중앙은행은 하루짜리 초단기 금리를 조절해서 전체 금리 커브를 조절함 but 초단기 금리가 0%면 어떻게 해야하지?<br>
>기준금리가 0%에 가까워 지자, 유동성을 공급하기 위해 정책금리 이외의 다른 방법을 강구해야 했다,<br>
>공개 조작의 범위를 확대하여 '유통시장'에서 단기국채뿐만 아니라 '장기국채'와 'MBS'를 매입해서 금리를 낮추고 유동성 공급<br>
>장기국채를 대부분 갖고있던 플레이어가 대형은행이므로, 은행의 위기에서 매우 효과적<br> 

나의첫금리공부참고
>1. 단순히 돈을 뿌리는게 아님 그것은 물가상승압력만 가져올 뿐이다 <br>
>2. 유통시장에서 국채와 모기지 채권을 매입하여<br>
>> * 은행이 예금인출에 대비한 금액을 중앙은행에 초과지준금에 대한 약간의 이자를 주고 모두 예치시킨 돈을 이용함<br>
>> * 이때 중앙은행이 화폐를 윤전기를 돌려서 을 동원하는 것이 아니라고 한다 (더블체크필요)
> 3. 채권 매도자에게 유동성을 부여하여 자본을 확충해주고, 모기지 채권 발행사들을 보증해줌 
>> * 이과정을 지속적으로 반복하여, 채권을 갖고있는 대형 은행들에게 유동성 공급
> 4. 평상시에 이런 행동은 인플레이션과 바로 연결되지만 신용경색으로 유동성이 축소되는 때에는 디플레이션(자산가격의 하락)을 둔화시키고 시장안정 효과를 가져옴 <br>
>> * 가장 큰 효과는 장기 채권금리의 하락하고(위험하락 + 장기간 금리가 낮을것이라는 기대) 이를 통해 대출이자 부담이 경감 <br>
>> * 재정정책과 연동될 수 있음
>> * 투자자들을 채권시장(금리하락)에서 주식시장으로 이동시키고 대출을 유도한다.
> 5. 중앙은행은 그 많은 국채를 어떻게 처리하는가?
>> * 경제침체이후 시중은행이 다시 가져가는 예치금은 지준에 대한 이자율을 높여서 적당히 메꾼다
>> * 채권은 시간이 지나 만기가 도래하므로 자연스레 그 규모가 줄어들 것이며 낮은 초과지준이자 (0.25%)로 자금을 빌려 3-4% 채권에 투자하므로 이자수익과 자본수익이 발생함.  대략 3조7천억달러나 점
>> 따라서 양적완화는 단기 금리가 낮고 장리 금리가 높을때 시도될 수 있음
>> 실제 ECB에 경우는 장기 금리가 낮을때 양적완화가 시행되어 한계가 있었음
